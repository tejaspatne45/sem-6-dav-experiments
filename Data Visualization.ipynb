```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(41)
```

```python
#Creating dataset
N_points=10000
n_bins=20
```

```python
#Creating distribution
x=np.random.randn(N_points)
```

```python
fig, ax=plt.subplots(figsize=(12,5))
ax.hist(x, bins=n_bins)
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(12,5))
sns.distplot(x, kde=True, color="orange")
```

```python
#Creating dataset
data={'C':20,'C++':25, 'Java': 30,'Python':35}
courses=list(data.keys())
values=list(data.values())
```

```python
plt.bar(courses, values, color='maroon')
plt.xlabel('courses')
plt.ylabel('values')
plt.title("Student enrollment in courses")
plt.show()
```
